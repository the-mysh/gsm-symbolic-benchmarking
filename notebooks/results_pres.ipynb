### When model fails to answer

```python
failed_answers = results_20x50_main.full_data[np.isnan(results_20x50_main.full_data['predicted_numerical_result'])]
failed_answers
```

```python
failed_answers['model'].value_counts()
```

```python
p = (100 * failed_answers['model'].value_counts() / results_20x50_main.full_data['model'].value_counts()).dropna()
p.name = "failed %"
p
```
